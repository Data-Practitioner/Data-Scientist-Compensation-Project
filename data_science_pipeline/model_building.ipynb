## 1. Multiple Linear Regression

```python
# Defining X & Y variable.
dataset = ml_data
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
# Training the model.
regressor = LinearRegression()
regressor.fit(X, y)
```

```python
# Predicting the dependent variable.
y_pred = regressor.predict(X)
np.set_printoptions(precision=2)
ml_pred = pd.DataFrame(np.concatenate((y_pred.reshape(len(y_pred),1), y.reshape(len(y),1)),1))
ml_pred
```

```python
# Evaluting algorithm performance.
r_square_mlr = r2_score(y, y_pred)
rmse_mlr = sqrt(mean_squared_error(y, y_pred))
mse_mlr = mean_squared_error(y, y_pred)
mae_mlr = mean_absolute_error(y, y_pred)
mape_mlr = mean_absolute_percentage_error(y, y_pred)

print("r2: ", r_square_mlr)
print("rmse: ", rmse_mlr)
print("mse: ", mse_mlr)
print("mae: ", mae_mlr)
print("mape: ", mape_mlr)
```
